```python
import pandas as pd
```

```python
data= pd.read_csv('/content/drive/MyDrive/DLC-IDS/Datasets/train_merged_attacks.csv')
data_test = pd.read_csv('/content/drive/MyDrive/DLC-IDS/Datasets/test_merged_attacks.csv')
```

```python
attack_types_to_replace = ['Probe', 'Dos','U2R']
replacement_attack_type = 'Other'
for attack_type in attack_types_to_replace:
    data['label'] = data['label'].replace(attack_type, replacement_attack_type)
    data_test['label'] = data_test['label'].replace(attack_type, replacement_attack_type)


# number of attack labels for train data
print(data['label'].value_counts())
# number of attack labels for test data
print(data_test['label'].value_counts())
```

```python
data.to_csv('/content/drive/MyDrive/DLC-IDS/Datasets/train_merged_R2L.csv', index = False)
data_test.to_csv('/content/drive/MyDrive/DLC-IDS/Datasets/test_merged_R2L.csv', index=False)
```
