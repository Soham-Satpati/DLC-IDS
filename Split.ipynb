```python
merged_data = merged_data.drop_duplicates()
```

```python
# Shuffle the dataset randomly
merged_data = merged_data.sample(frac=1, random_state=42) # Set a random seed for reproducibility

# Calculate the split indices
total_rows = len(merged_data)
train_size = int(0.8 * total_rows)
test_size = total_rows - train_size

# Split the dataset into three parts
train_data = merged_data[:train_size]
test_data = merged_data[train_size:train_size + test_size]

# Save the datasets to the specified location
train_data.to_csv('/content/drive/MyDrive/DLC-IDS Revision (NSL-KDD)/80-20 Split/Datasets/NSL_KDD_80.csv', index=False)
test_data.to_csv('/content/drive/MyDrive/DLC-IDS Revision (NSL-KDD)/80-20 Split/Datasets/NSL_KDD_20.csv', index=False)
```
